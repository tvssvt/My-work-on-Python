```python
#Underfitting and overfitting
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_absolute_error as mae 

def get_mae(max_leaf_nodes,trainx,valx,trainy,valy):
    model = DecisionTreeRegressor(max_leaf_nodes=max_leaf_nodes,random_state = 0)
    model.fit(trainx,trainy)
    pvalu = model.predict(valx)
    MAE = mae(valy, pvalu)
    return MAE
```

```python
for max_leaf_nodes in [5,50,500,5000]:
    e = get_mae(max_leaf_nodes,trainx,valx,trainy,valy)
    print("max leaf nodes: %d \t\t mean absolute error: %d" %(max_leaf_nodes,e))
```

```python
#optimal leaves would be 500
```

```python
# Random Forests Model
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
fm = RandomForestRegressor(random_state = 0)
fm.fit(trainx,trainy)
pmm = fm.predict(valx)
fmae = mae(valy, pmm)
print(fmae)
```

```python
# When we compare errors in both models, we can see that random forest model is more optimal then decision tree.
```
